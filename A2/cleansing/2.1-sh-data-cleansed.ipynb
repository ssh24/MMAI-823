```python
# use >0.8 cut off for highly correlated features, and drop them (either returns or excessReturnFromRiskModel)
# no highly anti-correlated features, so we leave everything else as is
df_train = df_train.drop('returns', axis=1)
# drop year column
df_train = df_train.drop('year', axis=1)
```

```python
# get all the train data columns
list(df_train.columns)
```

```python
# use >0.8 cut off for highly correlated features, and drop them (either returns or excessReturnFromRiskModel)
# no highly anti-correlated features, so we leave everything else as is
df_test = df_test.drop('returns', axis=1)
# drop year column
df_test = df_test.drop('year', axis=1)
```

```python
# get all the test data columns
list(df_test.columns)
```

```python
# Andrews work: drop the >6sd things we saw in data exploration from train 
returns = df_train.loc[:,'outputReturn']

Z = (returns-np.mean(returns))/np.std(returns)

indexA = Z>=-6 
indexB = Z<=6
index = indexA & indexB

df_train = df_train.loc[index,:]
```

```python
## Andrews work: split test/train X/Y
x_train = df_train.drop('outputReturn',axis =1)
y_train = df_train.loc[:,['outputReturn']]

x_test = df_test.drop('outputReturn',axis =1)
y_test = df_test.loc[:,['outputReturn']]

## use standard scaler to normalalize
scaler = preprocessing.StandardScaler().fit(x_train)

x_train_norm = scaler.transform(x_train)
x_test_norm = scaler.transform(x_test)

x_train_names = pd.DataFrame(x_train_norm, columns = x_train.columns)
x_test_names = pd.DataFrame(x_test_norm, columns = x_test.columns)
```

```python
output_train = pd.concat([x_train_names.reset_index(drop=True),y_train.reset_index(drop=True)],axis =1)
output_test = pd.concat([x_test_names.reset_index(drop=True),y_test.reset_index(drop=True)],axis =1)
```
